```python
from scipy.io import loadmat
import matplotlib.pyplot as plt

for p in paths[:5]:
    mat = loadmat(p)
    img, ri = mat2npy(mat)
    
    calib = calibration(img, ri)
    gau_img = gaussian_3d(img)
    cal_gau = gaussian_3d(calib)
    
    plt.figure(figsize=(20, 10))
    plt.title("1. Origin, 2. After calibration, 3. Origin + Gaussian, 4. Calib + Gaussian")
    plt.subplot(1,4,1)
    plt.imshow(img[:, :, img.shape[2]//2])
    plt.subplot(1,4, 2)
    plt.imshow(calib[:, :, img.shape[2]//2])
    plt.subplot(1,4,3)
    plt.imshow(gau_img[:, :, img.shape[2]//2])
    plt.subplot(1,4,4)
    plt.imshow(cal_gau[:, :, img.shape[2]//2])
    
```
