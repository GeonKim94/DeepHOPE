```python
import numpy as np
import time

def pad_image_concatenate(image, pad_x, pad_y, pad_value=0):
    padded_image = np.concatenate(
        (np.full((image.shape[0], image.shape[1], pad_x), pad_value),
         image,
         np.full((image.shape[0], image.shape[1], pad_x), pad_value)),
        axis=2
    )

    padded_image = np.concatenate(
        (np.full((image.shape[0], pad_y, padded_image.shape[2]), pad_value),
         padded_image,
         np.full((image.shape[0], pad_y, padded_image.shape[2]), pad_value)),
        axis=1
    )

    return padded_image

def pad_image_numpy_pad(image, pad_x, pad_y, pad_value=0):
    pad_width = ((0, 0), (pad_y, pad_y), (pad_x, pad_x))
    padded_image = np.pad(image, pad_width, mode='constant', constant_values=pad_value)
    return padded_image

# Create a sample 3D image
original_image = np.random.randint(0, 65535, size=(1024, 1024, 12), dtype=np.uint16)

# Padding parameters
pad_x = 512
pad_y = 512
pad_value = 100

# Measure time for numpy.concatenate
start_time = time.time()
concatenated_image = pad_image_concatenate(original_image, pad_x, pad_y, pad_value)
concatenate_time = time.time() - start_time

# Measure time for numpy.pad
start_time = time.time()
padded_image = pad_image_numpy_pad(original_image, pad_x, pad_y, pad_value)
numpy_pad_time = time.time() - start_time

# Verify correctness
assert np.array_equal(concatenated_image, padded_image)

# Print results
print(f"Time taken using numpy.concatenate: {concatenate_time:.6f} seconds")
print(f"Time taken using numpy.pad: {numpy_pad_time:.6f} seconds")
```
